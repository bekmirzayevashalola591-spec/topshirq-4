```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([92,90,99,95,91])
mylabels=["Chiziqli algebra","Kiberxavfsizlik asoslari","Diskret tuzilmalari","Elektronika va sxemalar","Suniy intellekt"]
mycolors=["red","yellow","blue","black","pink"]
myexplode=[0.2,0,0.5,0,0]
plt.pie(x,labels=mylabels,startangle=90,explode=myexplode,colors=mycolors)
plt.show()
```
